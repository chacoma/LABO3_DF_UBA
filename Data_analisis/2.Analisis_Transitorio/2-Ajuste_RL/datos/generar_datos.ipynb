```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## RL

```python
arx = "RL.txt"

df1 = pd.read_csv(arx, sep='\s+', header=None, names=['time', 'Vin','VL' ,'I'])

print(df1.dtypes)

df1.head()
```

```python
t1 = df1["time"]

Vin1 = df1["Vin"].to_numpy() + np.random.normal(size=len(t1))*0.011

VL1 = df1["VL"].to_numpy() + np.random.normal(size=len(t1))*0.011

I1 = 1*df1["I"].to_numpy() + np.random.normal(size=len(t1))*0.00001

fig,ax = plt.subplots()

ax.plot( t1, Vin1 )
ax.plot( t1, VL1 )

#ax.plot( t1, I1 )

plt.show()



fig,ax = plt.subplots()
ax.plot( t1, I1 )
plt.show()
```

```python
data = { "time":t1, "Vin":Vin1, "VL":VL1, "I":I1 }

df = pd.DataFrame(data=data)

df.to_csv("data_RL.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def modelo(f, R,L,C):

    w = 2*np.pi*f

    T = 1.0/ np.sqrt( 1 + ((w*L - 1.0/(w*C))**2)/R**2  )

    return T


def modelo_corregido(f, R,L,C, c):

    w = 2*np.pi*f
   
    f0 = (1/(2*np.pi))*(1/np.sqrt(L*C))
    f1 = f0/1.1
    f2 = 1.1*f0

    X = w*L - 1/(w*C) 

    Re = R+ c*(f-f0) 


    T = ( 1/ np.sqrt( 1 + (X/Re)**2 ) ) 

    return T




R = 100
L = 10e-3
C = 400e-12

c = 0.025


f0 = (1/(2*np.pi))*(1/np.sqrt(L*C))
f1 = f0/1.05
f2 = 1.05*f0


freqs = np.linspace( f1,f2 , 1000 )


fig,ax = plt.subplots()

ax.plot( freqs, modelo(freqs, R,L,C), linestyle='--', alpha=0.5 )

ax.plot( freqs, modelo_corregido(freqs, R,L,C, c) , c='red')


ax.axvline(f0, linestyle='--', alpha=0.3, c='black')

plt.show()


```

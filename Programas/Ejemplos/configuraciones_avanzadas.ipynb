## OSCILOSCOPIO

#### Mostrar y traer resta de canales


```python
# 1. Definir la función MATH como la resta de CH1 menos CH2
osc.write('MATH:DEFINE "CH1 - CH2"') 

# 2. Encender la visualización de la forma de onda MATH
osc.write("MATH:DISPLAY ON")

# 3. Obtener los datos de la forma de onda MATH para procesarlos en Python
osc.write("DATA:SOURCE MATH") # Seleccionar MATH como la fuente de datos
math_data = osc.query("CURVE?") # Pedir los datos de la curva
```

#### visualizar ch1 vs ch2, 

```python
# 1. Configurar la visualización al modo X-Y (Canal 1 = Eje X, Canal 2 = Eje Y)
# Este comando es para la visualización de los canales, no de la función MATH.
osc.write("DISplay:FORmat XY") 

# 2. Encender los canales para que se visualicen en los ejes X e Y
osc.write("CH1:STATE ON")
osc.write("CH2:STATE ON")

# 3. Ajustar escalas vertical y horizontal
osc.write("CH1:SCALE 0.2")    # Escala del eje X 
osc.write("CH2:SCALE 0.05")   # Escala del eje Y 
```

#### Promediado de la señal

```python
# Establecer la cantidad de ciclos a promediar (ejemplo: 64)
osc.write("ACQuire:NUMAVg 64") 

# Activar el modo de adquisición de Promedio
osc.write("ACQuire:MODe AVERage")

# El osciloscopio ahora mostrará el resultado promediado. 
# Puedes adquirir los datos una vez que el promedio se complete.
#El promedio es una configuración global que afecta a la forma en que se capturan todos los datos, 
# por lo que no es necesario especificar el canal. Simplemente asegúrate de que 
# el canal que te interesa (CH1, CH2, o MATH) esté activado (CH<x>:STATE ON) 
# para que se muestre el resultado promediado.
# el averge se muestra en el osc y se adquire en los datos

# para volver al modo normal
osc.write("ACQuire:MODe SAMple")

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Barrido de tensi√≥n

```python
df = pd.read_csv('datos/barrido_tension.txt', 
                 sep='\t|,',  
                 engine='python',  
                 skiprows=1, 
                 names=['Vf', 'Vd', 'I'],
                 dtype='float'
)

df.head()
```

```python
#plot
fig,ax = plt.subplots(figsize=(4,3))

ax.scatter( df["Vd"],-df["I"] , label="")

ax.set_xlabel("$V_d$ [Hz]")
ax.set_ylabel("$I$ ")
ax.legend()
plt.show()
```

### Rampa

```python
df2 = pd.read_csv('datos/rampa.txt', 
                 sep='\t|,',  
                 engine='python',  
                 skiprows=0, 
                 dtype='float'
)

df2.head()
```

```python
R=100

#plot
fig,ax = plt.subplots(figsize=(4,3))

ax.plot( df2["time"],df2["Vf"] , label="V fuente")

ax.plot( df2["time"],df2["VR"] , label="V resist")

ax.plot( df2["time"],df2["Vd"] , label="V diodo")

ax.set_xlabel("$t$ [s]")
ax.set_ylabel("$V$ [V]")
ax.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## RLC subamortiguado

```python
arx = "RLC_sub.txt"

df = pd.read_csv(arx, sep='\s+', header=None, skiprows=1, names=['time', 'Vin','VC' ,'VL'])

print(df.dtypes)

df.head()

df.plot( x="time", y=["Vin","VC"],kind='line', figsize=(12,3) )
```

```python
df["Vin"]= df["Vin"] + np.random.normal(size=df.shape[0])*0.025

df["VC"] = df["VC"] + np.random.normal(size=df.shape[0])*0.025

df["VL"]  = df["VL"] + np.random.normal(size=df.shape[0])*0.025

df.plot( x="time", y=["Vin", "VC"],kind='line', alpha=0.75, figsize=(12,4) )
```

```python
df.to_csv("data_RLC_sub.csv", index=False)
```

### RLC sobreamortiguado

```python
arx = "RLC_sobre.txt"

df2 = pd.read_csv(arx, sep='\s+', header=None, skiprows=1, names=['time', 'Vin','VC' ,'I'])

print(df2.dtypes)

df2.head()
```

```python
df2.plot( x="time", y=["Vin","VC"],kind='line', figsize=(12,4) )
```

```python
df2["Vin"]= df2["Vin"] + np.random.normal(size=df2.shape[0])*0.01

df2["VC"] = df2["VC"] + np.random.normal(size=df2.shape[0])*0.01

df2["I"]  = df2["I"] + np.random.normal(size=df2.shape[0])*0.000001

df2.plot( x="time", y=["Vin", "VC"],kind='line', alpha=0.5, figsize=(12,4) )

df2.plot( x="time", y=["I"],kind='line', alpha=0.5, figsize=(12,3) )
```

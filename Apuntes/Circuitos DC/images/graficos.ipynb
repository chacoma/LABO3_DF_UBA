```python
import numpy as np
import matplotlib.pyplot as plt
import plotter_lib as pl

x= np.arange(0.0,10,0.01)
y = 1/(1+x)**2
y2 = x/(1+x)**2

plot = pl.Plotter(H=2.5,W=3.5)

plot.ax.plot(x,y, c="C2", label='$P$')
plot.ax.plot(x,y2, c="C3", label="$P^\prime$")
plot.ax.scatter([1],[0.25], zorder=3, c="C5")
ann = "$(1, 1/4)$"
plot.ax.annotate( ann,(1.2,0.3) )
plot.ax.set_ylabel("$P/P_0$")
plot.ax.set_xlabel("$x=R'/R$")
plot.show("potencias.png")
```

```python
cmap = plt.colormaps['Greens'].reversed() # Cambia 'viridis' por otra paleta
colors = cmap(np.linspace(0, 1, 10))  # Obtiene 10 colores de la paleta


x= np.arange(0.01,10, 0.01)
f = lambda u,a : a/(1+a) - u/(1+u)


plot = pl.Plotter(H=2.5,W=3.5)

i=0
labels=[r"$\alpha=0.5$",r"$\alpha=1$",r"$\alpha=2$"]

for a in [0.5,1,2]:
    plot.ax.plot(x,f(x,a), c=colors[i], label=labels[i])

    i+=1

plot.ax.axhline( 0, linestyle='--',c='black', alpha=0.2 )

plot.ax.set_ylabel( "$V_{AB}/V_0$" )
plot.ax.set_xlabel( "$x = R_x/R_2$" )

plot.show("Vab_puenteW.png")
```

```python
plot = pl.Plotter(H=2.5,W=3.5)

x1= np.arange(-1.,1., 0.01)
f1 = lambda u : (1+u)/(2+u)   
plot.ax.plot(x1,f1(x1), c="C0", linewidth=2, label="Real")


x2= np.arange(-1.,1., 0.01)
f2 = lambda u : (1/2)+(1/4)*u  
plot.ax.plot(x2,f2(x1), c="red", label="Approx.", zorder=4)

#x1= np.arange(-0.5,0.5, 0.01)
#plot.ax.plot(x1,f2(x1)-f1(x1), c="black", linestyle="--", alpha=0.5, zorder=0, label="$\delta$")


plot.ax.axhline( 0.5, linestyle='--',c='black', alpha=0.2 )
plot.ax.axvline( 0, linestyle='--',c='black', alpha=0.2 )

plot.ax.set_yticks([0,0.25,0.5,0.75])

plot.ax.set_ylabel( "$V_{L}/V_0$" )
plot.ax.set_xlabel( "$\Delta R/R$" )

plot.show("divisor_plot.jpg", loc=4)
```

```python
import sympy as sp

# Definir la variable y la funcin
x = sp.symbols('x')
f = (1 + x) / (1 + x / 2)**2

# Calcular la primera derivada
dfdx = sp.diff(f, x)

# Encontrar los puntos crticos
critical_points = sp.solve(dfdx, x)
critical_points
```

```python
# Calcular la segunda derivada
d2fdx2 = sp.diff(dfdx, x)

# Evaluar la segunda derivada en x = 0
second_derivative_at_0 = d2fdx2.subs(x, 0)
second_derivative_at_0
```

```python
# Expansin de Taylor de primer orden en x = 0
taylor_expansion = f.series(x, 0, 2).removeO()
taylor_expansion
```
